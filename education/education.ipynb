```python
#coding utf8
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
education = pd.read_csv(r'/home/katia/Downloads/POP_XWAP_SEX_AGE_EDU_NB_Q-filtered-2022-03-25.csv')
education = education[['ref_area.label','classif2.label','time','obs_value' ]]
#prod = prod[['geo', 'TIME_PERIOD','OBS_VALUE']]
education
df_education = education.pivot_table('obs_value', index = ['ref_area.label'],columns = ['time', 'classif2.label'] )
df_education
```

```python
total_education = education[(education['classif2.label'] == 'Education (Aggregate levels): Total')]
total_education
total_education = total_education.pivot_table('obs_value', index = ['ref_area.label'],columns = ['time', 'classif2.label'] )
total_education['Total Average'] = total_education.iloc[:,:88].mean(axis=1, skipna=True)
total_education
```

```python
intermediate_education = education[(education['classif2.label'] == 'Education (Aggregate levels): Intermediate')]
intermediate_education
intermediate_education = intermediate_education.pivot_table('obs_value', index = ['ref_area.label'],columns = ['time', 'classif2.label'] )
intermediate_education['Intermidiate Average'] = intermediate_education.iloc[:,:88].mean(axis=1, skipna=True)
intermediate_education
```

```python
basic_education = education[(education['classif2.label'] == 'Education (Aggregate levels): Basic')]
basic_education
basic_education = basic_education.pivot_table('obs_value', index = ['ref_area.label'],columns = ['time', 'classif2.label'] )
basic_education['Basic Average'] = basic_education.iloc[:,:88].mean(axis=1, skipna=True)
basic_education
```

```python
advanced_education = education[(education['classif2.label'] == 'Education (Aggregate levels): Advanced')]
advanced_education = advanced_education.pivot_table('obs_value', index = ['ref_area.label'],columns = ['time', 'classif2.label'] )
advanced_education
advanced_education['Advanced Average'] = advanced_education.iloc[:,:88].mean(axis=1, skipna=True)
advanced_education
```

```python
aggregate = total_education[['Total Average']]
aggregate['basic Average'] = (basic_education['Basic Average'] / total_education['Total Average']) *100
aggregate['intermediate Average'] = (intermediate_education['Intermidiate Average'] / total_education['Total Average'])*100
aggregate['advanced Average'] = (advanced_education['Advanced Average'] / total_education['Total Average'])*100
aggregate
```
