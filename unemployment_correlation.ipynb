```python
#coding utf8
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


sentiment = pd.read_csv('./data/sentiment.csv') 
productivity = pd.read_csv('./data/GDP_per_quarter.csv') 
unem = pd.read_csv('./data/unem.csv')
unem.fillna(0)
productivity.fillna(0)

```

```python
#clean the dataset, keeping only the columns you need
clean = unem[['LOCATION', 'TIME', 'Value']]
clean
```

```python
df = clean.pivot_table('Value',  ['LOCATION'],'TIME')
df = df.rename({'AUT': 'Austria'})
df = df.rename({'BEL': 'Belgium'})
df = df.rename({'DEU': 'Germany'})
df = df.rename({'ESP': 'Spain'})
df = df.rename({'EST': 'Estonia'})
df = df.rename({'FIN': 'Finland'})
df = df.rename({'FRA': 'France'})
df = df.rename({'GRC': 'Greece'})
df = df.rename({'IRL': 'Ireland'})
df = df.rename({'ITA': 'Italy'})
df = df.rename({'LTU': 'Lithuania'})
df = df.rename({'LUX': 'Luxembourg'})
df = df.rename({'LVA': 'Latvia'})
df = df.rename({'PRT': 'Portugal'})
df = df.rename({'NLD': 'Netherlands'})
df = df.rename({'SVK': 'Slovakia'})
df = df.rename({'SVN': 'Slovenia'})

df.reset_index(inplace=True)
df
```

```python
df = df.reindex([16, 0, 1, 2, 4, 3, 5, 6, 7, 8, 9, 10, 11, 12, 13, 15, 17, 14])

arr = []
for name in df['LOCATION'].unique(): 
    for n in productivity['Country'].unique():
        if name == n:
            arr.append(name)
        

prod = productivity.drop(labels = [3, 10, 16], axis = 0)

prod.reset_index(inplace=True)
final = df.drop(labels = [14], axis = 0)
#final.reset_index(inplace= True)

```

```python
final['mean'] = final.iloc[:,1:89].mean(axis=1, skipna = True)
prod['mean'] = prod.iloc[:,2:90].mean(axis=1, skipna = True)
final.reset_index(inplace=True)
```
