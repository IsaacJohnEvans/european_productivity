```python
import pandas as pd
wages = pd.read_csv('./dataframes/wages_df.csv')
wages = wages.set_index('ref_area.label')
wages = wages.rename(columns = {'Average': 'Average Wages'})
wages
```

```python
unemployment = pd.read_csv('./dataframes/unemployment_df.csv')
unemployment = unemployment.set_index('ref_area.label')
unemployment
```

```python
# Correlation only with advanced education: 0.60
# Medium correlation with Intermediate education:-0.43
# No correlation with Basic education
education = pd.read_csv('./dataframes/education_df.csv')
education = education.set_index('TIME')
education
```

```python
inflation = pd.read_csv('./dataframes/inflation_df.csv')
inflation = inflation.set_index('LOCATION')
inflation
```

```python
depression = pd.read_csv('./dataframes/depression_df.csv')
depression = depression.set_index('TIME')
depression
```

```python
age_df = pd.read_csv('./dataframes/age_df.csv')
age_df = age_df.set_index('TIME')
age_df
```

```python
sleep = pd.read_csv('./dataframes/sleep_df.csv')
sleep = sleep.set_index('Country')
sleep
```

```python
df = unemployment
df['Advanced Education'] = education['Advanced Average']
df['Intermediate Education'] = education['intermediate Average']
df['Inflation'] = inflation['Average_Infilation']
df['depression'] = depression['Average Depression Rate']
df['% of young employers(15-24)'] = age_df['1524 Employers Average']
df['Average Wages'] = wages['Average Wages']
df['Average Sleep'] = sleep['Hours']


df
```
